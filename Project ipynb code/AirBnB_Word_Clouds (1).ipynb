## Word Clouds & Count

### Setup

```python
import pandas as pd
import os
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
```

### Imports

```python
clean_airbnb_df = pd.read_csv("AirBnB_Open_Data_Cleaned.csv", sep=",",low_memory=False)
```

```python
cleaned_airbnb_hr_df = pd.read_csv("AirBnB_Open_Data_House_Rules.csv", sep=",",low_memory=False)
```

### Word Cloud

```python
# Make sure its clean and all strings
clean_airbnb_df['NAME'] = clean_airbnb_df['NAME'].fillna('').astype(str)

# biiiiig string
text = ' '.join(clean_airbnb_df['NAME'])

#Adding default stopwords
stopwords = set(STOPWORDS)

# Generate the word cloud
wordcloud = WordCloud(width=1200, height=600, background_color='white', stopwords=stopwords).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Tokenize the NAME column and remove stopwords
clean_airbnb_df['TOKENIZED_NAME'] = clean_airbnb_df['NAME'].apply(lambda x: [word for word in word_tokenize(x.lower()) if word.isalpha() and word not in stopwords])

# Flatten the list of tokenized words and count the frequency of each word
all_words = [word for tokens in clean_airbnb_df['TOKENIZED_NAME'] for word in tokens]
word_freq = Counter(all_words)

# Get the top 20 most common words
top_20_words = word_freq.most_common(20)

# Display the top 20 words in a nice list
print("Top 20 words by frequency:")
for word, freq in top_20_words:
    print(f"{word}: {freq}")
```

```python
# Make sure its clean and string
cleaned_airbnb_hr_df['house_rules'] = cleaned_airbnb_hr_df['house_rules'].fillna('').astype(str)

# combine
text = ' '.join(cleaned_airbnb_hr_df['house_rules'])

wordcloud = WordCloud(width=1200, height=600, background_color='white', stopwords=stopwords).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
