### Setup

```python
import pandas as pd
import os
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

### Imports

```python
clean_airbnb_df = pd.read_csv("AirBnB_Open_Data_Cleaned.csv", sep=",",low_memory=False)
```

```python
cleaned_airbnb_hr_df = pd.read_csv("AirBnB_Open_Data_House_Rules.csv", sep=",",low_memory=False)
```

### Word Cloud

```python
# Make sure its clean and all strings
clean_airbnb_df['NAME'] = clean_airbnb_df['NAME'].fillna('').astype(str)

# biiiiig string
text = ' '.join(clean_airbnb_df['NAME'])

#Adding default stopwords
stopwords = set(STOPWORDS)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Make sure its clean and string
cleaned_airbnb_hr_df['house_rules'] = cleaned_airbnb_hr_df['house_rules'].fillna('').astype(str)

# combine
text = ' '.join(cleaned_airbnb_hr_df['house_rules'])

wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
